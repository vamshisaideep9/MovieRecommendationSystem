Lower case

```python
df['genres'] = df['genres'].str.lower()
```

```python
df['keywords'] = df['keywords'].str.lower()
df['title'] = df['title'].str.lower()
df['production'] = df['production'].str.lower()
df['tagline'] = df['tagline'].str.lower()
```

```python
df.head(3)
```

Remove punctuation

```python
import string

exclude = string.punctuation
```

```python
exclude
```

```python
import re
def remove_punc(text):
    if isinstance(text, str):  
        return re.sub(r'[^\w\s]', ' ', text)  # Removes all punctuation
    return text
```

```python
df = df.map(remove_punc)
```
